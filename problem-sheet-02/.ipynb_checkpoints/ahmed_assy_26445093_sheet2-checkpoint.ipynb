```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import griddata
```

# Exercise 2.1: algorithm runtimes.

```python

# Import the dataset and ignore comments
runtimes_df = pd.read_csv('runtimes.csv', comment='#')
runtimes_df = pd.melt(runtimes_df,id_vars=['algo','size'],var_name='threads',value_name='runtime')

runtimes_df['threads'] = runtimes_df['threads'].str.extract(r'time(\d)').astype(int)
runtimes_df = runtimes_df.dropna()
# Add new column called agorithm type which represent the label in the plot
runtimes_df['algorithm_type'] = runtimes_df.apply(lambda x: 'Single' if x['algo'] == 'single' else 'Distributed with ' + str(x['threads']) + ' Threads',axis = 1)

```

```python
# Line plot for the data with no lagend
sns.lineplot(data=runtimes_df, x='size', y='runtime', hue='algorithm_type', marker='o', legend=False)

# add direct labels to the right side of the graph with some offcests 
algorithm_types = runtimes_df['algorithm_type'].unique()
offsets = [[1000000, 0], [-100000, 100], [1000000, 500], [1000000, 400], [1000000, 200], [1000000, -100]]
for index, algorithm_type in enumerate(algorithm_types):
    subset_df = runtimes_df[runtimes_df['algorithm_type'] == algorithm_type]
    max_x = subset_df['size'].max()
    max_y = subset_df['runtime'].max()
    color = sns.color_palette()[list(algorithm_types).index(algorithm_type)]
    plt.text(max_x + offsets[index][0], max_y + offsets[index][1], algorithm_type, fontsize=10, va='center', ha='left', color=color)

# Remove the right and top lines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(True)
plt.gca().spines['left'].set_visible(True)

# Change the axes to logarithmic scale
plt.xscale('log')
plt.yscale('log')

# Add title and show
plt.title('Runtime VS Problem size \n for various numbers of worker threads')
plt.show()
```

```python
# Change the size type to be String so that i can use it as label with the same colors palette 
runtimes_df['size'] = runtimes_df['size'].astype(str)

# Line plot for the data with no lagend
sns.lineplot(data=runtimes_df[runtimes_df['algo'] == 'distributed'], x='threads', y='runtime', hue='size', marker='o', legend=False)

# add direct labels to the right side of the graph with some offcests 
sizes = runtimes_df['size'].unique()
offsets = [[0, -2], [0, -22], [0, -110], [0, -210], [0, -750], [0, -3500]]
for index, size in enumerate(sizes):
    subset_df = runtimes_df[runtimes_df['size'] == size]
    max_x = subset_df['threads'].max()
    max_y = subset_df['runtime'].max()
    color = sns.color_palette()[list(sizes).index(size)]
    plt.text(max_x + offsets[index][0], max_y + offsets[index][1], size, fontsize=10, va='center', ha='left', color=color)

# Remove the right and top lines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(True)
plt.gca().spines['left'].set_visible(True)

# Change the x.axis to logarithmic scale
plt.yscale('log')

# Add title and show
plt.title('Runtime VS number o f Threads \n for various problem sizes')

plt.show()
```

**In an ideal scenario, we would expect to see a decrease in runtime as the number of threads increases. However, in practice, we often encounter a point where further increasing the number of threads does not lead to significant reductions in runtime. This is typically due to overheads or hardware limitations. For instance, the difference in runtime when moving from one thread to two threads is usually much larger across all problem sizes compared to the difference observed when going from four threads to five threads. This discrepancy highlights the impact of hardware limitations, which ideally should not result in such disparities.**












# Exercise 2.2: precipitation data.

```python
# open the file with the correct encoding as it is not utf-8
with open("zehn_min_rr_Beschreibung_Stationen.txt", "r", encoding="latin1") as file:
    file_content = file.read()

# Split the content into lines and remove leading or trailing whitespaces
lines = [line.strip() for line in file_content.strip().split("\n")]

# Remove header separator line and split header into individual columns
header = lines[0].split()
header = [column.strip() for column in header]

# Initialize lists to store station information
station_ids = []
von_datums = []
bis_datums = []
stationshoehe = []
geoBreite = []
geoLaenge = []
stationsname = []
bundesland = []

# extract the data from each line after splitting it by space and ignoring the first two lines
for line in lines[2:]:  
    parts = line.split()
    station_ids.append(parts[0])
    von_datums.append(parts[1])
    bis_datums.append(parts[2])
    stationshoehe.append(parts[3])
    geoBreite.append(parts[4])
    geoLaenge.append(parts[5])
    stationsname.append(" ".join(parts[6:-1])) # as the station name may contains spaces 
    bundesland.append(parts[-1]) # as all the states contains - instead of spaces so we need only the last word

# Create a DataFrame to store station information
stations_df = pd.DataFrame({
    header[0]: station_ids,
    header[1]: von_datums,
    header[2]: bis_datums,
    header[3]: stationshoehe,
    header[4]: geoBreite,
    header[5]: geoLaenge,
    header[6]: stationsname,
    header[7]: bundesland
})

# Save the DataFrame to a CSV file
stations_df.to_csv("weather_stations.csv", index=False)
stations_df.head(5)
```

```python
stations_df['Stationshoehe'] = stations_df['Stationshoehe'].astype(int)
stations_df['geoBreite'] = stations_df['geoBreite'].astype(float)
stations_df['geoLaenge'] = stations_df['geoLaenge'].astype(float)
stations_df['Stations_id'] = stations_df['Stations_id'].astype('int64')

```

```python
plt.figure(figsize=(16, 9))
# scatter plot for the stations locations (Longitude and Latitude) vs the stations elevation
plt.scatter(stations_df['geoLaenge'], stations_df['geoBreite'], c=stations_df['Stationshoehe'])
plt.colorbar(label='Elevation (m)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Position of Weather Stations and Elevation')
plt.grid(True)
# to make the map more realestic as the number of Longitude is double the number of Latitude.
plt.gca().set_aspect(2)
plt.show()
```

```python
# read the cv and renam the id column to be the same for both datasets
ten_min_df = pd.read_csv("10min_processed.csv")
ten_min_df.rename(columns={'stationid': 'Stations_id'}, inplace=True)

# convert the date formate
ten_min_df['date'] = pd.to_datetime(ten_min_df['date'], format='%Y%m%d%H%M')

# replace missing values encoded as '-999' to zero
ten_min_df['rain'] = ten_min_df['rain'].apply(lambda x: max(x, 0))

# Add hours column and group by the hours
ten_min_df['hour'] = ten_min_df['date'].dt.hour
grouped_df = ten_min_df.groupby(['Stations_id', 'hour'])

# Sum the rain values within each hour
hourly_precipitation = grouped_df['rain'].sum().reset_index()

# merge the two data frames to get the locations of each station from the first data frame
hourly_precipitation_with_coords = hourly_precipitation.merge(stations_df[['Stations_id', 'geoBreite', 'geoLaenge']], how='left', on='Stations_id')


```

```python

# create subplots with 6 rows and 4 columns for all 24 hours, increase figsize and reduce spacing
fig, axs = plt.subplots(nrows=4, ncols=6, figsize=(20, 30), sharex=True, sharey=True, gridspec_kw={'hspace': 0.2, 'wspace': 0.15})

# plot the map for the hourly rain precipitation
for hour in range(24):
    # get the rain for each hour
    hourly_precipitation_selected = hourly_precipitation_with_coords[hourly_precipitation_with_coords['hour'] == hour]

    ax = axs.flatten()[hour]

    # scatter plot for the stations locations (Longitude and Latitude) vs the hourly rain precipitation
    scatter = ax.scatter(hourly_precipitation_selected['geoLaenge'], hourly_precipitation_selected['geoBreite'], c=hourly_precipitation_selected['rain'], cmap='viridis')
    ax.set_title(f'At {(str(hour) + ":00 AM") if hour < 12 else (str(hour - 12) + ":00 PM")}')
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    ax.grid(True)
    ax.set_aspect(2)  # To make the map more realistic as the number of Longitude is double the number of Latitude

# Add overall title and adjust the position
fig.suptitle('Geographical Position of Weather Stations and Hourly Rain Precipitation on 20/04/2024', fontsize=16, y=0.9)

# Add colorbar and adjust the position
plt.colorbar(scatter, ax=axs, orientation='horizontal', label='Rain precipitation (mm)', aspect=50, pad=0.03)

plt.show()
```

```python
# Load the griddata.npz file
data = np.load('griddata.npz')
geolat = data['geolat']
geolong = data['geolong']
ind = data['ind']
```

```python

# create subplots with 6 rows and 4 columns for all 24 hours, increase figsize and reduce spacing
fig, axs = plt.subplots(nrows=4, ncols=6, figsize=(20, 30), sharex=True, sharey=True, gridspec_kw={'hspace': 0.2, 'wspace': 0.15})

# Plot interpolated precipitation data for each hour
for hour in range(24):
    ax = axs.flatten()[hour]
    
    # get the rain for each hour
    hourly_data = hourly_precipitation_with_coords[hourly_precipitation_with_coords['hour'] == hour]
    
    # Interpolate precipitation data to the grid
    interpolated_rain = griddata((hourly_data['geoLaenge'], hourly_data['geoBreite']), hourly_data['rain'], (geolong, geolat), method='linear')    
    
    # Mask out regions where 'ind' is False as it is not part of the map
    interpolated_rain = np.ma.masked_where(~ind, interpolated_rain)
    
    # Plot the image
    im = ax.imshow(interpolated_rain, extent=(geolong.min(), geolong.max(), geolat.min(), geolat.max()), aspect='auto',origin = 'lower')
    ax.set_title(f'At {(str(hour) + ":00 AM") if hour < 12 else (str(hour - 12) + ":00 PM")}')
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    ax.set_aspect(2)  # To make the map more realistic as the number of Longitude is double the number of Latitude  

# Add overall title and adjust the position
fig.suptitle('Geographical Position of Weather Stations and Hourly Rain Precipitation on 20/04/2024', fontsize=16, y=0.9)

# Add colorbar
plt.colorbar(scatter, ax=axs, orientation='horizontal', label='Rain precipitation (mm)', aspect=50, pad=0.05)

# Adjust layout
plt.show()
```
